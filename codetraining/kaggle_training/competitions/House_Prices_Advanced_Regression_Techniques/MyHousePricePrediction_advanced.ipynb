```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
# print(os.getcwd())#获得当前工作目录
train_df = pd.read_csv("./input/train.csv",index_col=0)
test_df = pd.read_csv("./input/test.csv",index_col=0)
```

```python
y_train = np.log1p(train_df.pop("SalePrice"))
```

```python
all_df =  pd.concat((train_df,test_df),axis=0)
```

```python
all_dummy_df = pd.get_dummies(all_df)
all_dummy_df.head()
```

```python
all_dummy_df.isnull().sum().sort_values(ascending=False).head()
```

```python
mean_cols = all_dummy_df.mean()
mean_cols.head()
```

```python
all_dummy_df = all_dummy_df.fillna(mean_cols)
```

```python
numeric_cols = all_df.columns[all_df.dtypes!="object"]
numeric_cols
```

```python
numeric_cols_means = all_dummy_df.loc[:,numeric_cols].mean()
numeric_col_std = all_dummy_df.loc[:,numeric_cols].std()
all_dummy_df.loc[:,numeric_cols] = (all_dummy_df.loc[:,numeric_cols]-numeric_cols_means)/numeric_col_std
```

```python
all_dummy_df.head(10)
```

```python
dummy_train_df = all_dummy_df.loc[train_df.index]
dummy_test_df = all_dummy_df.loc[test_df.index]
# train_df.index
```

```python
dummy_train_df.shape,dummy_test_df.shape
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(15)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import cross_val_score
```

```python
X_train = dummy_train_df.values
X_test = dummy_test_df.values
params = [1,10,15,20,25,30,40]
test_scores = []
for param in params:
    clf = BaggingRegressor(n_estimators=param,base_estimator=ridge)
    test_score = np.sqrt(-cross_val_score(clf,X_train,y_train,cv=10,scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(params,test_scores)
plt.title("n_estimator vc CV Error")
```

```python
bag = BaggingRegressor(n_estimators=25,base_estimator=ridge)
bag.fit(X_train, y_train)
y_bag = np.expm1(bag.predict(X_test))
submission_df = pd.DataFrame(data= {'Id' : test_df.index, 'SalePrice': y_bag})
submission_df.to_csv("sample_submission.csv",index=0)
```

```python
params = [10, 15, 20, 25, 30, 40, 50, 60, 70, 100]
test_scores = []
for param in params:
    clf = BaggingRegressor(n_estimators=param)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(params, test_scores)
plt.title("n_estimator vs CV Error");
```
